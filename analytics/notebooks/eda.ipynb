```python
import pandas as pd
from pathlib import Path
```

```python
base = Path('analytics/exports')
users = pd.read_csv(base / 'users.csv')
accounts = pd.read_csv(base / 'accounts.csv')
transactions = pd.read_csv(base / 'transactions.csv')
postings = pd.read_csv(base / 'postings.csv')
users.head()
```

```python
transactions['timestamp'] = pd.to_datetime(transactions['timestamp'])
transactions['date'] = transactions['timestamp'].dt.date
kpis = transactions.groupby('date').agg(total_tx=('id', 'count'), total_amount=('amount', 'sum'))
kpis.tail()
```

```python
segment = transactions.groupby('account_id').agg(tx_count=('id', 'count'), volume=('amount', 'sum'))
segment = segment.merge(accounts[['id', 'user_id']], left_on='account_id', right_on='id')
segment.groupby('user_id').sum().sort_values('volume', ascending=False).head(10)
```
